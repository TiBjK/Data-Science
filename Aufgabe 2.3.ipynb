```python
import time
import numpy as np

# Gruppenbericht
#Mitglieder [ Noah Bennardo, Tim Kern, Michael Schäfer, Vincent Wallek ]

groessen = [10_000, 100_000, 1_000_000, 5_000_000]

print("Vergleich: Python-Liste vs. NumPy-Array")
print("Operation: Alle Elemente quadrieren\n")

for size in groessen:
    print(f"► Test mit {size:,} Elementen".replace(",", "."))
    print("-" * 55)

    # a) + b) Daten erzeugen
    liste = list(range(size))
    array = np.arange(size)

    # Python-Liste
    start = time.time()
    quadrat_liste = [x*x for x in liste]
    zeit_liste = time.time() - start

    # NumPy
    start = time.time()
    quadrat_array = array ** 2
    zeit_array = time.time() - start

    print(f"Python-Liste :  {zeit_liste:.5f} s")
    print(f"NumPy-Array  :  {zeit_array:.5f} s")

    
    if zeit_array < 0.00001:                               # quasi sofort fertig
        print("NumPy war SO schnell, dass wir es kaum messen konnten! (> 1000× schneller)\n")
    else:
        schneller = zeit_liste / zeit_array
        print(f"→ NumPy war ca. {schneller:.0f}× schneller\n")

print("Fazit der Gruppe:")
print("• NumPy ist bei großen Datenmengen HUNDERTE Male schneller.")
print("• Normale Python-Listen werden ab ca. 100.000 Elementen richtig langsam.")
print("• Deshalb: Immer NumPy für Mathe mit vielen Zahlen benutzen!")
print("")
print("Wir haben den Code jetzt extra robust gemacht – kein Absturz mehr ")
```
