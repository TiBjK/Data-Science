## Aufgabe 2.3

```python
import time
import numpy as np

größen = [10_000, 100_000, 1_000_000, 5_000_000]

for size in größen:
    # Python-Liste
    liste = list(range(size))
    start = time.time()
    quadrat_liste = [x ** 2 for x in liste]
    dauer_liste = time.time() - start
    
    # NumPy-Array
    array = np.arange(size)
    start = time.time()
    quadrat_array = array ** 2
    dauer_array = time.time() - start
    
    print(f"{size:,} Elemente".replace(",", "."))
    print(f"Python-Liste : {dauer_liste:.5f} s")
    print(f"NumPy-Array  : {dauer_array:.5f} s")
    
    if dauer_array < 0.00001:
        print("NumPy extrem schnell (kaum messbar)\n")
    else:
        faktor = dauer_liste / dauer_array
        print(f"NumPy ca. {faktor:.0f}× schneller\n")
```

### Fazit
NumPy ist bei numerischen Operationen auf großen Datenmengen deutlich schneller als reine Python-Listen. Der Vorteil entsteht durch vektorisierte Berechnungen, die intern in optimiertem C-Code ablaufen – ohne langsame Python-Schleifen.
Je größer die Datenmenge, desto stärker der Performance-Vorsprung:
bei mehreren Millionen Elementen ist NumPy oft 20- bis 50-mal schneller oder mehr.
