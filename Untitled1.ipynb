```python
from scipy.stats.contingency import association


ct = pd.crosstab(df["GraduateOrNot"], df["TravelInsurance"])
print(ct)
# Berechnen und Ausgabe von Cramers V
print('Cramers V zwischen Graduate or Not und Travel Insurance: ', association(ct, method="cramer"))

ct = pd.crosstab(df['EverTravelledAbroad'], df['FrequentFlyer'])
print(ct)
print('Cramers V zwischen EverTravelledAbroad und FrequentFlyer: ', association(ct, method="cramer"))


ct = pd.crosstab(df['FrequentFlyer'], df['TravelInsurance'])
print(ct)
print('Cramers V zwischen Frequent Flyer und Travel Insurance: ', association(ct, method="cramer"))
```
