```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# --- 1. Filter: Alter 2035 ---
df_filtered = df[(df['Age'] >= 20) & (df['Age'] <= 35)].copy()

# --- 2. Altersgruppen ---
bins = [20, 25, 30, 35, 36]
labels = ['2024', '2529', '3034', '35']
df_filtered['AgeGroup'] = pd.cut(
    df_filtered['Age'], bins=bins, labels=labels,
    include_lowest=True, right=False
)

# --- 3. Einkommensgruppen ---
df_filtered['IncomeGroup'] = df_filtered['AnnualIncome'].apply(
    lambda x: '>120k' if x > 120000 else '120k'
)

# --- 4. Abschlussrate berechnen ---
rate = df_filtered.groupby(['AgeGroup', 'IncomeGroup'])['TravelInsurance'].mean() * 100

# --- WICHTIG: Beide Spalten erzwingen! ---
rate = rate.unstack(fill_value=0)  # Normales unstack
# Falls Spalten fehlen  hinzufgen
for col in ['120k', '>120k']:
    if col not in rate.columns:
        rate[col] = 0.0

# --- 5. Alle Altersgruppen erzwingen ---
rate = rate.reindex(labels, fill_value=0)

# --- 6. Plot: Nebeneinanderliegende Balken ---
fig, ax = plt.subplots(figsize=(10, 6))
bar_width = 0.35
x = np.arange(len(labels))

# Balken
ax.bar(x - bar_width/2, rate['120k'], bar_width,
       label='120k', color='lightcoral', edgecolor='black')
ax.bar(x + bar_width/2, rate['>120k'], bar_width,
       label='>120k', color='lightgreen', edgecolor='black')

# Prozentwerte auf Balken
for i, age in enumerate(labels):
    low = rate.loc[age, '120k']
    high = rate.loc[age, '>120k']
    ax.text(i - bar_width/2, low + 1, f'{low:.0f}%', ha='center', va='bottom', fontweight='bold')
    ax.text(i + bar_width/2, high + 1, f'{high:.0f}%', ha='center', va='bottom', fontweight='bold')

# Beschriftung
ax.set_xlabel('Altersgruppe', fontsize=12)
ax.set_ylabel('Abschlussrate Reiseversicherung (%)', fontsize=12)
ax.set_title('Vergleich: Abschlussrate nach Einkommen und Alter\n(2035 Jahre)', fontsize=14, pad=20)
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.set_ylim(0, 100)
ax.legend(loc='upper left', bbox_to_anchor=(1, 1))
ax.grid(axis='y', alpha=0.3, linestyle='--')

plt.tight_layout()
plt.show()
```
