```python
#Indexing

import numpy as np

# Array erzeugen: 1 bis 100 als (10, 10)
arr = np.arange(1, 101).reshape(10, 10)
print("Originales Array:")
print(arr)

# a) 1., 5. und letzte Zeile
print("\na) Zeilen 1, 5 und letzte:")
arr = np.arange(1, 101).reshape(10, 10)
print(arr[[0, 4, 9]])

# b) 3. und 7. Spalte in ungeraden Zeilen
print("\nb) 3. und 7. Spalte in ungeraden Zeilen:")
arr = np.arange(1, 101).reshape(10, 10)
print(arr[1::2, [2, 6]])

# c) 3x3 Subarray links oben
print("\nc) 3x3 links oben:")
arr = np.arange(1, 101).reshape(10, 10)
print(arr[:3, :3])

# d) 2. und 4. Zeile negativ machen
arr = np.arange(1, 101).reshape(10, 10)
arr[[1, 3]] = -arr[[1, 3]]
print("\nd) Nach Negativ-Machen der 2. und 4. Zeile:")
print(arr)

# e) Alle Werte < 25 durch 0 ersetzen
arr = np.arange(1, 101).reshape(10, 10)
arr[arr < 25] = 0
print("\ne) Nach Ersetzen aller < 25 durch 0:")
print(arr)
```
