# 1 Business Understanding & Literaturrecherche

## Datensatz
**Travel Insurance Prediction Data** (Kaggle, 2020)  
Link: https://www.kaggle.com/datasets/tejashvi14/travel-insurance-prediction-data  

Der Datensatz enthält 1.987 (davon viele Duplikate) Kunden einer Reiseversicherung mit folgenden Merkmalen:  
Age, Employment Type, GraduateOrNot, AnnualIncome, FamilyMembers, ChronicDiseases, FrequentFlyer, EverTravelledAbroad sowie die Zielvariable **TravelInsurance** (0 = nicht gekauft, 1 = gekauft).

## Formulierte Forschungsfragen & wichtigste Erkenntnisse der Papers

| Paper | Forschungsfrage | Wichtigste Prädiktoren | Beste Modelle & Performance | Erkenntnisse |
|-------|------------------|-------------------------|-----------------------------|--------------|
| Rubi et al. (2022) | Welche Kundemerkmale beeinflussen den Kauf einer Reiseversicherung am stärksten? | 1. EverTravelledAbroad<br>2. AnnualIncome<br>3. FrequentFlyer | Random Forest & Logistic Regression (Accuracy ~0.85) | Auslandserfahrung ist der mit Abstand stärkste Treiber |
| Lim et al. (2023) | Wie hat sich das Kaufverhalten nach COVID-19 verändert und welche ML-Methode performt am besten? | 1. EverTravelledAbroad<br>2. AnnualIncome<br>3. Employment Type | K-Nearest Neighbors (Accuracy 0.82, F1 0.80, AUC 0.78) – überraschend besser als RF | Nach-Pandemie-Effekte sind im Datensatz nicht stark sichtbar |

**Gemeinsame Erkenntnisse beider Studien:**  
- Kunden, die bereits ins Ausland gereist sind, kaufen mit >80 % Wahrscheinlichkeit eine Versicherung.  
- Höheres Jahreseinkommen korreliert stark positiv mit dem Kauf.  
- Klassische Modelle erreichen 0.80–0.85 Accuracy, aber es bleibt Potenzial nach oben.
