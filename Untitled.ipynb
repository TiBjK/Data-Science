```python
b = np.array(range(5))
c = np.array(range(3,8))

d = b + c
e = b - c
f = b*c
g = np.mean(b)
h = np.mean(c)
b[b < np.mean(b)] = 0
c[c < np.mean(c)] = 0


print(d)
print(e)
print(f)
print(g)
print(h)
print(b)
print(c)
```
